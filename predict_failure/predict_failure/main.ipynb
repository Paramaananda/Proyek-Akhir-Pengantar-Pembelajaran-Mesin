```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from imblearn.under_sampling import RandomUnderSampler
from sklearn.feature_selection import r_regression
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("train.csv")
```

```python
df.info()
```

```python
df
```

```python
df = df.drop(columns=["id"])
```

```python
# df = df.dropna(axis=0)
```

```python
df.info()
```

```python
for column in df.columns[:-1]:
    if (df[column].dtype == "object"):
        le = LabelEncoder()
        df[column] = pd.Series(le.fit_transform(df[column]))
```

```python
df.info()
```

```python
imputer = SimpleImputer(missing_values=np.nan)
```

```python
df[df.columns[:-1]] = imputer.fit_transform(df[df.columns[:-1]])
```

```python
df.info()
```

```python
df
```

```python
x = df[df.columns[:-1]]
y = df[df.columns[-1]]
```

```python
selector = r_regression(x,y)
```

```python
print(selector)
```

```python
selector = [abs(x) for x in selector]
```

```python
min(selector)
```

```python
max(selector)
```

```python
selected_feature = []

for corr in selector:
    if(corr<0.1):
        selected_feature.append(False)
    else:
        selected_feature.append(True)
```

```python
rus = RandomUnderSampler(random_state=0)
x, y = rus.fit_resample(x, y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
hasil_rf = model.predict(x_test)
```

```python
print(accuracy_score(y_test,hasil_rf))
```
