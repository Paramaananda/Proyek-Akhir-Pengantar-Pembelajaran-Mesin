```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from imblearn.under_sampling import RandomUnderSampler
from sklearn.feature_selection import r_regression
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
df = pd.read_csv("train.csv")
```

```python
df = df.drop(columns=["id",'product_code'])
```

```python
for column in df.columns[:-1]:
    if (df[column].dtype == "object"):
        le = LabelEncoder()
        df[column] = pd.Series(le.fit_transform(df[column]))
```

```python
imputer = SimpleImputer(missing_values=np.nan)
```

```python
df[df.columns[:-1]] = imputer.fit_transform(df[df.columns[:-1]])
```

```python
x = df[df.columns[:-1]]
y = df[df.columns[-1]]
```

```python
selector = r_regression(x,y)
```

```python
print(selector)
```

```python
selector = [abs(x) for x in selector]
```

```python
min(selector)
```

```python
max(selector)
```

```python
selected_feature = []

for corr in selector:
    if(corr<0.1):
        selected_feature.append(False)
    else:
        selected_feature.append(True)
```

```python
rus = RandomUnderSampler(random_state=0)
x, y = rus.fit_resample(x, y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
knn = KNeighborsClassifier()
nb = GaussianNB()
dt = DecisionTreeClassifier()
svm = SVC()
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
```

```python
knn.fit(x_train,y_train)
nb.fit(x_train,y_train)
dt.fit(x_train,y_train)
svm.fit(x_train,y_train)
rf.fit(x_train,y_train)
gb.fit(x_train,y_train)
```

```python
knn_predict = knn.predict(x_test)
nb_predict = nb.predict(x_test)
dt_predict = dt.predict(x_test)
svm_predict = svm.predict(x_test)
rf_predict = rf.predict(x_test)
gb_predict = gb.predict(x_test)
```

```python
print("knn accuracy : ")
print(accuracy_score(y_test,knn_predict))
```

```python
print("nb accuracy : ")
print(accuracy_score(y_test,nb_predict))
```

```python
print("dt accuracy : ")
print(accuracy_score(y_test,dt_predict))
```

```python
print("svm accuracy : ")
print(accuracy_score(y_test,svm_predict))
```

```python
print("rf accuracy : ")
print(accuracy_score(y_test,rf_predict))
```

```python
print("gb accuracy : ")
print(accuracy_score(y_test,gb_predict))
```

```python
knn_predict = knn_predict.tolist()
nb_predict = nb_predict.tolist()
dt_predict = dt_predict.tolist()
svm_predict = svm_predict.tolist()
rf_predict = rf_predict.tolist()
gb_predict = gb_predict.tolist()
```

```python
final_result = []
for i in range (len(knn_predict)):
    temp = [knn_predict[i],nb_predict[i],dt_predict[i],svm_predict[i],rf_predict[i],gb_predict[i]]
    counter_1 = temp.count(1)
    counter_0 = temp.count(0)
    if (max(temp) == counter_0):
        final_result.append(0)
    else:
        final_result.append(1)
```

```python
final_result = pd.Series(final_result)
print('final accuracy: ')
print(accuracy_score(y_test,final_result))
```
