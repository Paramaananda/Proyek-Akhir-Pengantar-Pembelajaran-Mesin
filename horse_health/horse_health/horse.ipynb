```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("train.csv")
dataset.head()
```

```python
dataset = dataset.drop(columns="hospital_number")
```

```python
dataset = dataset.drop(columns="id")
```

```python
for column in dataset.columns:
    if dataset[column].dtype == "float64" or dataset[column].dtype == "int64":
        dataset[column].fillna(value=dataset[column].mean(), inplace=True)
```

```python
# dataset = dataset.dropna()
```

```python
dataset.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
x = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
label_encoder = LabelEncoder()
for column in x.columns:
    # Periksa tipe data kolom, lalu lakukan label encoding jika tipe data adalah 'object' (kategorikal)
    if x[column].dtype == 'object':
        x[column] = label_encoder.fit_transform(x[column])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Akurasi:", accuracy)
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
```
