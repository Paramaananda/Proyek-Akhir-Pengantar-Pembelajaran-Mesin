```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=["id","height(cm)","hearing(left)","hearing(right)","Urine protein","ALT",])
```

```python
df.info()
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
```

```python
svm_model = SVC()
```

```python
x = df[df.columns[:-1]]
y = df[df.columns[-1]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
knn_model.fit(x_train,y_train)
```

```python
svm_model.fit(x_train,y_train)
```

```python
rf_model = RandomForestClassifier()
```

```python
rf_model.fit(x_train,y_train)
```

```python
hasil_knn = knn_model.predict(x_test)
```

```python
hasil_svm = svm_model.predict(x_test)
```

```python
hasil_rf = rf_model.predict(x_test)
```

```python
print("akurasi knn:")
print(accuracy_score(y_test,hasil_knn))
```

```python
print("akurasi svm:")
print(accuracy_score(y_test,hasil_svm))
```

```python
print("akurasi random forest:")
print(accuracy_score(y_test,hasil_rf))
```
